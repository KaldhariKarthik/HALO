```python
import sys
from pathlib import Path
import torch

# Add YOLOv5 path to sys
yolov5_path = Path('C:/Users/karth/Desktop/UPSKILL/HALO/PHASE1/yolov5')
sys.path.append(str(yolov5_path))

from train import run

def train_yolo():
    data_yaml = yolov5_path / 'glare.yaml'  # nc:1, names:['class0']
    weights = 'yolov5s.pt'
    epochs = 50
    batch_size = 8
    imgsz = 416
    device = '0' if torch.cuda.is_available() else 'cpu'

    print(f"Training on device: {device}")

    sys.argv = [
        'train.py',
        '--data', str(data_yaml),
        '--weights', weights,
        '--epochs', str(epochs),
        '--batch-size', str(batch_size),
        '--imgsz', str(imgsz),
        '--device', device,
        '--project', 'runs/train',
        '--name', 'glare_class0_augmented',
        '--exist-ok',
        '--single-cls',  # single class mode

        # Aggressive augmentation flags
        '--augment',  # enable augmentation explicitly
        '--hsv-h', '0.05',    # increase hue augmentation
        '--hsv-s', '1.0',     # increase saturation augmentation
        '--hsv-v', '0.7',     # increase value augmentation
        '--degrees', '20',    # rotate images up to 20 degrees
        '--translate', '0.3', # translate images by 30%
        '--scale', '1.2',     # scale images up to 120%
        '--shear', '10',      # shear by 10 degrees
        '--flipud', '0.5',    # 50% chance to flip upside down
        '--fliplr', '0.8',    # 80% chance to flip left-right
        '--mosaic', '1.0',    # enable mosaic (default is 1.0)
        '--mixup', '0.5'      # enable mixup with 50% probability
    ]

    run()

train_yolo()
```
